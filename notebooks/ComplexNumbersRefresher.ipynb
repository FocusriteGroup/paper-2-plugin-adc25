```python
import numpy as np
import matplotlib.pyplot as plt
```

## Multiply two complex numbers

```python
a = 1 + 2j
b = 3 - 3j

# remember that 1j * 1j = -1
print(1j * 1j)

#   a * b 
# = (1 + 2j)(3 - 3j) 
# = 3 - 3j + 6j + 6 
# = 9 + 3j

c = a * b
print(c)
```

## Draw a single complex number in the plane

```python
a = 1 + 2j
b = 3 - 3j

plt.figure(6)
plt.clf()
plt.scatter(a.real, a.imag, marker='x')
ax = plt.gca()
ax.set_xlim([-4, 4])
ax.set_ylim([-4, 4])
plt.grid()
plt.show()
```

## Decompose two complex numbers into phase-angle and magnitude and compare the result of these methods

```python
a = 1 +1j
b = -1 + 2j
c = a * b

# use arctan2 to find the phase angle

angle_a = np.arctan2(a.imag, a.real)
magnitude_a = abs(a)

angle_b = np.arctan2(b.imag, b.real)
magnitude_b = abs(b)

angle_c = angle_a + angle_b
if (angle_c > np.pi):
    angle_c = angle_c - 2*np.pi

print(angle_c)
print(np.arctan2(c.imag, c.real))
```

## Use a complex exponential to place complex numbers on the unit circle and place them in the phase plane

```python
theta = np.pi / 4
c = np.exp(1j * theta)

# Use arctan2 to get back the angle theta we started with and prove that this did what we expected.
print(np.arctan2(c.imag, c.real)) # pi / 4 ~= 0.78

# The magnitude of the complex exponential is always 1
print(abs(c)) # 1.0
```

## Draw a circle using complex exponentials

```python
# set up 
sample_rate = 4800 # samples / second
frequency = 4.0 # Hz (cycles / second)

omega = 2 * np.pi * (frequency / sample_rate) # radians / sample
rotation = np.exp(1j * omega)

N = sample_rate

points = np.zeros(N + 1, dtype=np.complex128)

points[0] = 1.0 
for i in range(1,len(points)):
    points[i] = points[i-1] * rotation

plt.figure(6)
plt.clf()
plt.plot(np.real(points), np.imag(points))
ax = plt.gca()
ax.set_xlim([-2, 2])
ax.set_ylim([-2, 2])
plt.grid()
plt.show()
```

## Draw a circle using eulers formula

```python
N = 32
a = np.cos(2 * np.pi / N) # real part
b = np.sin(2 * np.pi / N) # imaginary part
rn = 1
cn = 0

points = np.zeros(N + 1, dtype=np.complex128)

points[0] = rn + cn * 1j
for i in range(1,len(points)):
    cn_ = a * cn - b * rn
    rn = b * cn + a * rn
    cn = cn_
    points[i] = rn + cn * 1j
    
plt.figure(6)
plt.clf()
plt.scatter(np.real(points), np.imag(points), marker='x')
ax = plt.gca()
ax.set_xlim([-2, 2])
ax.set_ylim([-2, 2])
plt.grid()
plt.show()
```
