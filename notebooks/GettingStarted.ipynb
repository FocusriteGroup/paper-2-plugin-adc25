# Getting Started with Jupyter Notebook

Code in Jupyter Notebook is divided into cells.

You can run a cell by first selecting it and then using the `run` button above or typing Shift + Return. 

This will run the cell and advance to the next cell.

```python
# Load the numpy and matplotlib libraries
import numpy as np
import matplotlib.pyplot as plt
```

Numpy works on arrays, and it works fastest if you create the array before you need it. 

One easy way to do this is with the `zeros` function. Which creates an array of a given length.

```python
N = 10
signal = np.zeros(N)
```

Once we have an array we can use `matplotlib` to plot the array. 

```python
plt.plot(signal)
plt.grid()
plt.show()
```

Other ways to create arrays include `ones` which generates an array of all 1s, and `arange` which can generate a sequence from a starting value to a stopping value with a given step-size, defaulting to 1.

```python
a = np.arange(1,16) # 1,2,3 ...,15
```

We can multiply vectors of the same size together element-by-element with a single function, multiply every element by a scalar value and add scalar values to every element.

```python
a = a / 16.0        # 0.0625, 0.125, 0.1875 ..., 0.9375
a = a * a           # square every element 
a += 1              # add 1 to every element
plt.plot(a)
plt.grid()
```
