# Filter Analysis

## The Impulse Signal

We begin with an impulse signal. The mathematical definition is the following:

$\displaystyle \delta(n) = \left\{ {1,\;n=0}\atop{0,\;n\neq 0.} \right.$

This signal is used to characterised LTI systems (this include our filter). By using it as input to our filter we can see its response over time. This response is therefore named impulse response. A stable filter will have an pulse response decaying to zero over time whereas a response growing towards infinity indicates an unstable filter.

````python
plt.figure()
s = plt.stem(impulse_signal(10))
plt.grid()
````

## Filter Impulse Response

We can use the impulse to get a basic characterisation of the behaviour of the filter by processing the impulse signal with the filter.

````python
plt.figure()
s = plt.plot(get_impulse_response(mvmfilter(440, 0.002, 10000), 100))
plt.grid()
plt.title("Impulse Response")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
````
